```python
years = list(range(2022, 2020, -1))
```

```python
all_matches = []
```

```python
standings_url = "https://fbref.com/en/comps/9/2021-2022/2021-2022-Premier-League-Stats"
```

```python
import time

for year in years:
  data = requests.get(standings_url)
  soup = BeautifulSoup(data.text)
  standing_table = soup.select('table.stats_table')[0]

  links = [l.get("href") for l in standings_table.find_all("a")]
  links = [l for l in links if "/squads/" in l]
  team_urls = [f"https://fbref.com{l}" for l in links]

  previous_season = soup.select("a.prev")[0].get("href")
  standings_url = f"https://fbref.com{previous_season}"

  for team_url in team_urls:
    team_name = team_url.split("/")[-1].replace("Stats", "").replace("-", " ")

    data = requests.get(team_url)
    matches = pd.read_html(data.text, match = "Scores & Fixtures")[0]

    soup = BeautifulSoup(data.text)
    links = [l.get("href") for l in soup.find_all("a")]
    links = [l for l in links if l and "all_comps/shooting/" in l]
    data = requests.get(f"https://fbref.com{links[0]}")
    shooting = pd.read_html(data.text, match = "Shooting")[0]
    shooting.columns = shooting.columns.droplevel()

    try:
      team_data = matches.merge(shooting[["Date", "Sh", "SoT","Dist", "Dist", "FK", "PK", "PKatt"]], on  = "Date")
    except ValueError:
      continue

    team_data = team_data[team_data["Comp"] == "Premier League"]
    team_data["Season"] = year
    team_data["Team"] = team_name

    all_matches.append(team_data)

    time.sleep(5)
```

```python
match_df = pd.concat(all_matches)

match_df.columns = [c.lower() for c in match_df.columns]
```

Predicting EPL matches


```python
import pandas as pd
```

```python
matches = pd.read_csv("matches.csv", index_col = 0)
```

```python
matches.head()
```

```python
matches.dtypes
```

```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches['venue_code'] = matches['venue'].astype("category").cat.codes
```

```python
matches['opp_code'] = matches['opponent'].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex = True).astype("int")
```

```python
matches["daycode"] = matches["date"].dt.dayofweek
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 50, min_samples_split = 10, random_state = 1)
```

```python
train = matches[matches["date"] < "2022-01-01"]
```

```python
test = matches[matches['date'] >= "2022-01-01"]
```

```python
predictors = ["venue_code", "opp_code", "hour", "daycode"]
```

```python
model.fit(train[predictors], train["target"])
```

```python
preds = model.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test['target'], preds)
accuracy
```

```python
combined = pd.DataFrame(dict(actual = test['target'],predicted = preds))
```

```python
pd.crosstab(index = combined['actual'], columns = combined['predicted'])
```

```python
from sklearn.metrics import precision_score
```

```python
precision = precision_score(test['target'], preds)
precision
```

Rolling average for better trend analysis

```python
def rolling(group, cols, new_cols):
  group = group.sort_values('date')
  rolling_stat = group[cols].rolling(3, closed = 'left').mean()
  group[new_cols] = rolling_stat
  group = group.dropna(subset = new_cols)
  return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'fk', 'dist', 'pk', 'pkatt']
new_cols = [f"{c}_rolling" for c in cols]
```

```python
matches_rolling = matches.groupby('team').apply(lambda x: rolling(x,cols,new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
  train = data[data["date"] < "2022-01-01"]
  test = data[data["date"] >= "2022-01-01"]
  model.fit(train[predictors], train['target'])
  preds = model.predict(test[predictors])
  combined = pd.DataFrame(dict(actual = test['target'],predicted = preds, index = test.index))
  precision = precision_score(test['target'], preds)
  return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
